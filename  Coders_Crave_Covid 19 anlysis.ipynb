```python
top_countries=data.groupby('Country/Region')['Confirmed'].max().nlargest(10)
```

```python
def build_graph(y_label):
    import matplotlib.pyplot as plt
    plt.figure(figsize=(20,8))
    return sns.barplot(x="Country/Region", y=y_label, data=data[:20])
```

```python
data["Deaths"] = (data["Deaths"]/data["Confirmed"])*100
data["Recovered"] = (data["Recovered"]/data["Confirmed"])*100
data['Active']=(data['Active']/data['Confirmed'])*100
```
